```python
# 去除文件名中 中文字符、 时间戳
import os
import re
import shutil

def remove_chinese(text):
    """移除字符串中的中文字符（Unicode范围：\u4e00-\u9fff）"""
    return re.sub(r'[\u4e00-\u9fff]', '', text)

def remove_timestamp(text):
    """
    移除时间戳（支持多种格式）：
    - 连续14位数字（如20250716213445）
    - 日期格式（YYYY-MM-DD, YYYY_MM_DD, YYYY/MM/DD等）
    - 时间格式（HH:MM:SS, HH:MM）
    - 日期时间组合（YYYY-MM-DD HH:MM:SS）
    """
    # 定义时间戳正则模式（按优先级排序）
    patterns = [
        r'\d{4}[-_/\.]\d{1,2}[-_/\.]\d{1,2}[\s_]\d{1,2}:\d{2}(:\d{2})?',  # 日期+时间组合
        r'\d{14}',  # 连续14位数字
        r'\d{4}[-_/\.]\d{1,2}[-_/\.]\d{1,2}',  # 单独日期
        r'\d{1,2}:\d{2}(:\d{2})?'  # 单独时间
    ]
    # 按顺序替换所有匹配的时间戳
    for pattern in patterns:
        text = re.sub(pattern, '', text)
    return text

def clean_filename(filename):
    """清理文件名：先移除时间戳，再移除中文字符"""
    # 分离文件名和扩展名（保留扩展名不变）
    name, ext = os.path.splitext(filename)
    # 处理主文件名部分
    name = remove_timestamp(name)  # 先移除时间戳
    name = remove_chinese(name)    # 再移除中文字符
    # 清理多余分隔符（连续下划线替换为单个）
    name = re.sub(r'_+', '_', name)
    name = name.strip('_')  # 去除首尾残留下划线
    return name + ext  # 重新组合

# 为目标文件夹路径
file_path = rf'/NAS2/Data1/lbliao/Data/MXB/segment/0716/Otsu_output'

# 遍历并重命名文件
for file in os.listdir(file_path):
    original_path = os.path.join(file_path, file)
    if not os.path.isfile(original_path):
        continue  # 跳过目录

    new_filename = clean_filename(file)
    if new_filename == file:
        print(f"无需修改: {file}")
        continue

    new_path = os.path.join(file_path, new_filename)
    try:
        shutil.move(original_path, new_path)
        print(f"重命名成功: 【{file}】→【{new_filename}】")
    except Exception as e:
        print(f"重命名失败: {file} (错误: {str(e)})")

print("处理完成！")
```
